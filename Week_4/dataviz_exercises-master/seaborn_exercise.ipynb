## Seaborn Assignment

```python
# import pandas
import pandas as pd

# import seaborn
import seaborn as sns

# import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv',sep=',')
df.drop('#',axis=1,inplace=True)
```

```python
df.head()
```

### Task 1: Plot a regression line between attack and defense.

```python
# plot a regression line between attack and defense
sns.regplot(x='Attack',y='Defense',data=df)
```

### Task 2: Plot a regression line between attack and defense while displaying a distinction between the different generations. 

```python
#plot a regression line between attack and defense with a hue for the generation
sns.lmplot(x='Attack', y='Defense', data=df, hue='Generation')
```

### Task 3: Make a boxplot for each feature of dataframe (without Generation and Legendary features) in one plot.

```python
# Make a boxplot for each feature of df excluding genereation and legendary
df2 = df.drop(['Total','Generation','Legendary'],axis=1)
sns.boxplot(data=df2)
```

### Task 4: Plot a correlation heatmap of the pokemon dataframe (not including the features Generation and Legendary).

```python
# Plot a correlation heatmap of the pokemon dataframe (df) not including the features generation and legendary

plt.figure(figsize=(10,6))
sns.set_context('notebook',font_scale=1.0)

df3 = df.drop(['Name','Type 1','Type 2','Generation','Legendary'],axis=1)

df3_corr = df3.corr()
sns.heatmap(df3_corr, cmap='coolwarm', linecolor='white',linewidth=1, annot=True)

# df3_pt = df3.pivot_table(index='Type 1',columns='Type 2',values='Total')
# sns.heatmap(df3_pt, cmap='Blues', linecolor='white', linewidth=1)
```
