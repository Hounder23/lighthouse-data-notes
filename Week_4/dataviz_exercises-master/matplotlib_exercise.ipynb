## Matplotlib Assignment

```python
# import pandas
import pandas as pd

# import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('company_sales_data.csv')
```

```python
df.head()
```

### Task 1: Read the `total_profit` of all months and display it using a line plot.

```python
# Q1: Read Total profit of all months and show it using a line plot
plt.plot(df['month_number'], df['total_profit'])
```

### Task 2: Read all the different product sales data and display it using a multiline plot.

```python
# Q2: Read all the different product sales data and display it using a multiline plot
face_cream = df['facecream']
face_wash = df['facewash']
toothpaste = df['toothpaste']
bathing_soap = df['bathingsoap']
shampoo = df['shampoo']
moisturizer = df['moisturizer']

months_x = df['month_number']

print(plt.style.available)

plt.style.use('seaborn-v0_8-paper')

plt.plot(months_x, face_cream, label='Face Cream')
plt.plot(months_x, face_wash, label='Face Wash')
plt.plot(months_x, toothpaste, label='Toothpaste')
plt.plot(months_x, bathing_soap, label='Bathing Soap')
plt.plot(months_x, shampoo, label='Shampoo')
plt.plot(months_x, moisturizer, label='Moisturizer')

plt.xlabel('Month')
plt.ylabel('Sales units in number')
plt.title('Sales data')

plt.legend()
plt.tight_layout()
```

### Task 3: Read the `total_profit` of each month and display it using a histogram to find out which profit ranges are the most common.

```python
# Q3: Read the total profit of each month and show it using the histogram to see the most common profit ranges
total_profit = df['total_profit']
median_profit = total_profit.median()
profit_range = [150000, 200000, 250000, 300000, 350000, 400000, 450000, 500000, 550000]
print(plt.style.available)

plt.style.use('fivethirtyeight')
plt.hist(total_profit, profit_range, label='Amount',edgecolor='black')
plt.axvline(median_profit, color='r', label='Median Profit')
plt.xlabel('Profit Range')
plt.ylabel('Number of Months in Range')
plt.title('Profit Range Frequency')
plt.legend()
plt.tight_layout()


```
